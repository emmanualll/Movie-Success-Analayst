```python
print("Zero budget:", (df['budget'] == 0).sum())
print("Zero revenue:", (df['revenue'] == 0).sum())
```

```python
df = df[(df["budget"] != 0) & (df["revenue"] != 0)]
```

The velow coloum tells us the roi of the movies whivh is basically revenue - budget / revenue


```python
df["roi"] = (df["revenue"] - df["budget"]) / df["budget"]
df[["budget", "revenue", "roi"]].head()
```

```python
sns.histplot(df["roi"], bins = 50, kde = True)
plt.title("Distribution of ROI across Movies")
plt.xlabel("Return on Investment")
plt.show()
```

```python
clean_df = df[(df["budget"] > 0) & (df["revenue"]>0)]

clean_df["budget_crores"] = clean_df["budget"] / 1e7
clean_df["revenue_crores"] = clean_df["revenue"] / 1e7

sns.scatterplot(data=clean_df, x="budget_crores", y="revenue_crores", alpha=0.5)
plt.xscale("log")
plt.yscale("log")
plt.title("Budget and Revenue Comparison")
plt.xlabel("Budget")
plt.ylabel("revenue")

plt.show()
```
